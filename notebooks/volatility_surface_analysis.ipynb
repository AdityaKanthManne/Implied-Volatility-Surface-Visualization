```
# volatility_surface_analysis.ipynb

import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import os
from src.surface_utils import prepare_surface_data, interpolate_iv_surface

# Create output directory
os.makedirs("plots", exist_ok=True)

# Load and prepare data
df = prepare_surface_data(r"C:\Users\adity\OneDrive\Documents\GitHub\Implied-Volatility-Surface-Visualization\data\uso_options_data.csv")
X, Y, Z = interpolate_iv_surface(df, grid_size=50)

# Plot the 3D surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k', alpha=0.8)

ax.set_title("Implied Volatility Surface")
ax.set_xlabel("Strike Price")
ax.set_ylabel("Days to Expiry")
ax.set_zlabel("Implied Volatility")
fig.colorbar(surf, shrink=0.5, aspect=10)

plt.tight_layout()
plt.savefig(r"C:\Users\adity\OneDrive\Documents\GitHub\Implied-Volatility-Surface-Visualization\plots/implied_volatility_surface.png")
plt.show()
```
